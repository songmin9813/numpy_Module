```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#주피터 사용시 라인 안에서 그래프가 그려진다는 의미
```

```python
x=np.linspace(0,10,11)
y=x**2+x+2+np.random.randn(11)#약간의 노이즈를 줌
print(x)
print(y)
```

### 그래프 출력하기
- plot 함수(선 그래프), scatter(점 그래프), hist(히스토그램) 등 사용
  - 함수의 parameter 혹은 plt의 다른 함수로 그래프 형태 및 설정을 변경 가능
  - 기본적으로 x,y에 해당하는 값이 필요

```python
plt.plot(x,y) #x와 y의 값이 기본적으로 주어져야 함
#선 형태의 그래프
```

```python
plt.scatter(x,y)
```

### 그래프에 주석 추가
- x,y 축 및 타이틀

```python
plt.xlabel('X values') #X 레이블 이름
plt.ylabel('Y values') #Y 레이블 이름
plt.title('Hello plt!') # 타이틀 추가
plt.plot(x,y)
```

- gird 추가

```python
plt.grid(True) #그리드(점선) 추가
plt.plot(x,y)
```

- x,y축 범위 지정

```python
plt.xlim(0,100)
plt.ylim(0,150)
plt.plot(x,y)
```

### plot 함수 parameters
- 그래프의 형태에 대한 제어 가능
- [matplotlib 기본 사용 가이드](https://matplotlib.org/stable/tutorials/index.html)

- 그래프의 색상 변경

```python
plt.plot(x,y,'r')# rgb 가능
```

```python
plt.plot(x,y,'#ff00ff')#직접 값 입력 가능
```

- 그래프 선스타일 변경

```python
plt.plot(x,y,'-.')
```

```python
plt.plot(x,y,'g^')#g는 초록색, ^는 산모양
```

```python
plt.plot(x,y,':m')#땡땡이에 보라색. 스타일 속성 순서는 마음대로
```

```python
plt.plot(x,y,':m',linewidth=5)#땡땡이에 보라색에 두께는 5
```

- keyword parameter 이용하여 모든 속성 설정
  - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
  - color
  - linestyle
  - marker
  - markerfacecolor
  - markersize 등

```python
plt.plot(x,y,color='black',linestyle='--',marker='^',markerfacecolor='blue',markersize=6)
#선 색은 검정. 선 스타일은 찍찍. 마커는 산 표시. 마커 색은 파랑. 마커 크기는 6
```

### subplot로 여러 그래프 출력하기
- subplot 함수로 구획을 구별하여 각각의 subplot 그래프 출력

```python
plt.subplot(2,2,1)#2콤마 2의 구역 생성+1구역 지정(총 4개의 plot 생성)
plt.plot(x,y,'r')
plt.subplot(2,2,2)#2구역 지정
plt.plot(x,y,'g')
plt.subplot(2,2,3)#3구역 지정
plt.plot(x,y,'b')
plt.subplot(2,2,4)#4구역 지정
plt.plot(x,np.exp(x),'m')
```

### hist 함수 사용(히스토그램)
- 도수 분포 정보를 표현해줌
- histogram 생성
- bins로 histogram bar 개수 설정

```python
data=np.random.randint(1,100,size=200)
print(data)
```

```python
plt.hist(data,bins=10)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
#관련 함수는 뒤에서 지정해도 상관없을 듯
```

### 도즈언! 로또 번호 추첨기를 함수화시켜 만들어보기

```python
class lottery:
    def __init__(self):
        self.list=np.random.choice(np.arange(1,46),size=6,replace=False)
        #randint를 사용하게 되면 중복을 허용하게 됨
        return

a=lottery()
print(a.list)
```

### 몬테 카를로 법칙을 이용하여 원주율 구해보기

```python
points=np.random.rand(1000000,2)
np.sum(np.sum(points**2,axis=1)<1)
print(np.sum(np.sum(points**2,axis=1)<1)*4/1000000)
```
